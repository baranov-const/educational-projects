#### Исследование пропусков

Перед исследованием пропусков сохраним исходное количество строк в переменную **in_len**

```python
in_len = data.shape[0]
```

```python
# визуализируем пропуски, для этого напишем отдельную функцию
def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 16)
            .set_title('Процент пропусков в столбцах' + "\n", fontsize = 22, color = 'SteelBlue')    
        );
        plt.xlabel('Проценты', fontsize=18, color = 'SteelBlue')
        plt.ylabel('Столбцы', fontsize=18, color = 'SteelBlue')
        plt.show()
    except:
        print('пропусков не осталось или произошла ошибка в первой части функции ')

pass_value_barh(data)
```

1. Исследуем столбцы **name** и **genre**

Как можно заметить, если есть пропуск **name**, то он есть и в столбце **genre**. Кроме того, в таких строках ещё по три пропуска. Всего таких строк 2 (мало), поэтому удалим их. В таких строках слишком мало информации

2. Столбец **year_of_release**

```python
# смотрим пропуски
data[data['year_of_release'].isna()]
```

Причина наличия 269 пропусков неясна. Необходимо написать разработчикам интернет-магазина.

К сожалению, дата релиза игры связана с платформой. Поэтому заменить пропуск на значение этой же игры на другой платформе нельзя. Кроме того, некорректно будет заменять год на значение внутри имени: например, игра Madden NFL 2004. Эта игра вышла в 2003 году. Значит, число 2004 не обязательно коррелирует с годом. 

Посмотрим на процент таких строк относительно всего датасета:

```python
data[data['year_of_release'].isna()].shape[0] / in_len * 100
```

Таких строк всего 1,6%, поэтому удалим их

```python
data = data.dropna(subset=['year_of_release'])
```

3. Столбцы **critic_score**, **user_score**, **rating**

В этих столбцах больше всего пропусков. Причина, по которой они не были загружены неясна. Возможно, данные просто отсутствуют. Необходимо написать разработчикам.

Заменим пропуски в таких строках значением-маркером, например, -100. 

```python
col = ['critic_score', 'user_score', 'rating']
data[col] = data[col].fillna(-100)
```

3. Посмотрим на уникальные значение столбца **user_score**

Как можно заметить, тип данных в этом столбце - object. Однако неудобно работать со строками. Поэтому изменим на float. Кроме того, присутствует значение tbd. ### Портрет пользователя каждого региона

#### Самые популярные платформы (ТОП-5) в каждом регионе

Популярность будем описывать количеством продаж по регионам (столбцы **na_sales**, **eu_sales**, **jp_sales**)

```python
# составляем таблицу для NA, где будет сумма продаж по платформам
na_relevant = (relevant_data
               .groupby('platform')['na_sales']
               .agg({'sum'})
               .sort_values(by='sum', ascending=False) 
              )
# добавим столбец с процентами продаж
na_relevant['share'] = round(na_relevant['sum'] / na_relevant['sum'].sum()*100, 2)

# отбираем ТОП-5 
na_relevant = na_relevant.head().reset_index()
na_relevant
```

```python
# составляем таблицу для EU, где будет сумма продаж по платформам
eu_relevant = (relevant_data
               .groupby('platform')['eu_sales']
               .agg({'sum'})
               .sort_values(by='sum', ascending=False) 
              )
# добавим столбец с процентами продаж
eu_relevant['share'] = round(eu_relevant['sum'] / eu_relevant['sum'].sum()*100, 2)

# отбираем ТОП-5 
eu_relevant = eu_relevant.head().reset_index()
eu_relevant
```

```python
# составляем таблицу для JP, где будет сумма продаж по платформам
jp_relevant = (relevant_data
               .groupby('platform')['jp_sales']
               .agg({'sum'})
               .sort_values(by='sum', ascending=False) 
              )
# добавим столбец с процентами продаж
jp_relevant['share'] = round(jp_relevant['sum'] / jp_relevant['sum'].sum()*100, 2)

# отбираем ТОП-5 
jp_relevant = jp_relevant.head().reset_index()
jp_relevant
```

```python
# теперь визуализируем эти зависимости 
plt.figure(figsize=(7, 5))
ax1 = sns.barplot(y='share', x='platform', data = na_relevant, palette="cubehelix")
plt.xlabel('Платформа')
plt.ylabel('Проценты')
plt.title('ТОП-5 популярных платформ в NA')
plt.show()


plt.figure(figsize=(7, 5))
ax2 = sns.barplot(y='share', x='platform', data = eu_relevant, palette="rainbow")
plt.xlabel('Платформа')
plt.ylabel('Проценты')
plt.title('ТОП-5 популярных платформ в EU')
plt.show()

plt.figure(figsize=(7, 5))
ax2 = sns.barplot(y='share', x='platform', data = jp_relevant, palette="rocket")
plt.xlabel('Платформа')
plt.ylabel('Проценты')
plt.title('ТОП-5 популярных платформ в JP')
plt.show()
```

### Проверка гипотез

#### Гипотеза 1

**Формулировка:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Переформулируем запрос в нулевую и альтернативную гипотезу:
    
**h0:** Разница между средними пользовательскими рейтингами платформ Xbox One и PC равна нулю

**h1:** Разница между средними пользовательскими рейтингами платформ Xbox One и PC не равна нулю

**Проверка**:

```python
# сформируем столбцы
XOne_table = relevant_data.query('(user_score != -100) and (platform == "XOne")')['user_score']
PC_table = relevant_data.query('(user_score != -100) and (platform == "PC")')['user_score']

# выведем среднее выборочное
print(f'Среднее выборочное пользовательского рейтинга для XOne = {round(XOne_table.mean(),2)}')
print(f'Среднее выборочное пользовательского рейтинга для PC = {round(PC_table.mean(),2)}')

# проверим гипотезу
alpha = 0.05 # уровень стат. значимости

test_1 = st.ttest_ind(XOne_table, PC_table, equal_var=False)

print(f'p-value = {test_1.pvalue}')

if test_1.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** из анализа следует, что гипотеза о равенстве двух средних пользовательских рейтингов платформ Xbox One и PC не отвергается. Несмотря на то, что средние в выборках отличны, истинные средние скорее всего равны.

#### Гипотеза 2

**Формулировка:** Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Переформулируем запрос в нулевую и альтернативную гипотезу:
    
**h0:** Разница между средними пользовательскими рейтингами жанров Action и Sports равна нулю

**h1:** Разница между средними пользовательскими рейтингами жанров Action и Sports не равна нулю

**Проверка**:

```python
# сформируем столбцы
action_table_gen = relevant_data.query('(user_score != -100) and (genre == "Action")')['user_score']
sports_table_gen = relevant_data.query('(user_score != -100) and (genre == "Sports")')['user_score']

# выведем среднее выборочное
print(f'Среднее выборочное пользовательского рейтинга для action = {round(action_table_gen.mean(),2)}')
print(f'Среднее выборочное пользовательского рейтинга для sports = {round(sports_table_gen.mean(),2)}')

# проверим гипотезу
alpha = 0.05 # уровень стат. значимости

test_2 = st.ttest_ind(action_table_gen, sports_table_gen, equal_var=False)

print(f'p-value = {test_2.pvalue}')

if test_2.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** гипотеза о том, что истинное среднее пользовательского рейтинга для игр жанра Action равно жанру Sports отвергается. 
С большой долей вероятности средние пользовательские рейтинги имеют статистически значимые отличия.
